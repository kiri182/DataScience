**Hash Function**

```python
# Impoty a module of hash function
import hashlib
```

```python
# Prep for input data
input_1 = "alice"
input_2 = "アリス"
input_3 = "Alice"

# Specify an algorithm for hash function
sha256 = hashlib.sha256()

# Transform input data to byte string and then apply them to the hash function
sha256.update(input_1.encode())
sha256.update(input_2.encode())
sha256.update(input_3.encode())
```

```python
# Show the output
print(f"ダイジェスト値:{sha256.digest()}")
print(f"ハッシュ値:{sha256.hexdigest()}")
```

```python
# The most simplest way
sha256_data = hashlib.sha256(input_1.encode()).hexdigest()
print(sha256_data)
```

```python
# The byte size of hash value
print(sha256.digest_size)
```

```python
# The block size of hash value
print(sha256.block_size)
```

```python
# The name of hash algorithum
print(sha256.name)
```

```python
# How to copy the hash object
print(sha256.copy())
```

**Key Derivation Function**

```python
pbkdf2_hmac_hash = hashlib.pbkdf2_hmac('sha256', input_1.encode(), b"Alice", 10000)
```

```python
print(f"pbkdf2_hmacによるハッシュ化: {pbkdf2_hmac_hash.hex()}")
```

```python
pbkdf2_scrypt = hashlib.scrypt(input_2.encode(), salt=b'Alice', n=2, r=1, p=1)
print(f"pbkdf2_scryptによるハッシュ化: {pbkdf2_scrypt.hex()}")
```
