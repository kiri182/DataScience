```python
import pandas as pd
import numpy as np
from datetime import date, timedelta
import matplotlib.pyplot as plt
```

```python
#Create test data

lastdate = date.today()
vmin = -1
vmax = 1
rowcount = 200
tpl_colname = ['x']
colcount = len(tpl_colname)

data = np.random.uniform(vmin, vmax, (rowcount, colcount))
rdate = pd.date_range(lastdate - timedelta(days=rowcount), periods=rowcount)
df = pd.DataFrame(data, index=rdate, columns=tpl_colname)
```

```python
plt.figure(figsize=(30, 6))
plt.plot(df.index, df['x'], linestyle = "solid", label='x')
plt.xlabel('date')
plt.ylabel('value')
plt.xticks(rotation=90)
plt.grid(True)
plt.legend()
plt.show()
```

```python
n = [1, 5, 10]
df_lag = pd.DataFrame()
for i in n:
  df_lag['x_lag{}'.format(i)] = df['x'].shift(i)
```

```python
df_concat = pd.concat([df, df_lag], axis=1)
df = df_concat
```

```python
plt.figure(figsize=(30, 6))
for lag in n:
    plt.plot(df.index, df[f'x_lag{lag}'], label=f'x_lag{lag}')

plt.xlabel('date')
plt.ylabel('value')
plt.xticks(rotation=90)
plt.grid(True)
plt.legend()
plt.show()
```
